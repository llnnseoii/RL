```python
#cmost_config

import numpy as np

class CMOSTConfig:
    def __init__(self):
        # ==========================================
        # 1. Risk Settings (from Risk_Settings.m)
        # ==========================================
        # í™˜ì ê°œì¸ë³„ ìœ„í—˜ë„ ë¶„í¬ë¥¼ ë§Œë“¤ê¸° ìœ„í•œ ë°±ë¶„ìœ„ìˆ˜ (ì˜ˆì‹œê°’)
        # MATLABì˜ handles.Variables.Ind_Risk_Percentiles ì— í•´ë‹¹
        self.risk_percentiles = [0.1, 0.2, 0.3, 0.5, 1.0, 1.5, 2.0, 3.0, 5.0, 8.0, 10.0, 15.0]
        
        # ë³´ê°„ë²•ì„ í†µí•´ 500ê°œ êµ¬ê°„ì˜ ìœ„í—˜ë„ ë¶„í¬ ìƒì„± (AdjustRiskGraph í•¨ìˆ˜ ì´ì‹)
        self.individual_risk_dist = self._interpolate_risk(self.risk_percentiles)

        # ìì—°ì‚¬ íŒŒë¼ë¯¸í„° (NumberCrunching.mì˜ AgeProgression ë“± ì¶”ìƒí™”)
        self.prob_onset_base = 0.04      # ê¸°ë³¸ ìš©ì¢… ë°œìƒë¥ 
        self.prob_growth = 0.15          # ìš©ì¢… ì„±ì¥ í™•ë¥  (Small -> Large)
        self.prob_malignancy = 0.05      # ì•”í™” í™•ë¥  (Large -> Cancer)
        self.prob_cancer_progress = 0.4  # ì•” ê¸°ìˆ˜ ì§„í–‰ í™•ë¥ 

        # ==========================================
        # 2. Screening Settings (from Screening_Settings.m)
        # ==========================================
        # ë¯¼ê°ë„(Sensitivity): [P1, P2, P3, P4, P5, P6, Ca1, Ca2, Ca3, Ca4]
        # P1~P6: ìš©ì¢… í¬ê¸°ë³„ (1mm ~ >10mm), Ca1~Ca4: ì•” ë³‘ê¸°ë³„
        
        self.sensitivities = {
            # ëŒ€ì¥ë‚´ì‹œê²½ (Colonoscopy): ìš©ì¢… ë°œê²¬ìœ¨ì´ ë†’ìŒ + ì¹˜ë£Œ ê°€ëŠ¥
            'colonoscopy': np.array([0.70, 0.80, 0.90, 0.95, 0.98, 0.99, 0.95, 0.98, 0.99, 0.99]),
            
            # ë¶„ë³€ì í˜ˆê²€ì‚¬ (FOBT): ìš©ì¢… ë°œê²¬ìœ¨ ë‚®ìŒ, ì•” ë°œê²¬ìœ¨ì€ ì¤‘ê°„
            'fobt':        np.array([0.05, 0.05, 0.10, 0.15, 0.20, 0.25, 0.50, 0.60, 0.70, 0.80]),
            
            # Septin9 ë“± ë‹¤ë¥¸ ê²€ì‚¬ ì¶”ê°€ ê°€ëŠ¥
            'sept9':       np.array([0.01, 0.02, 0.05, 0.10, 0.15, 0.20, 0.60, 0.75, 0.85, 0.90])
        }

        # ë¹„ìš© ë° ë³´ìƒ ì„¤ì • (RL í•™ìŠµìš©)
        self.cost = {
            'wait': 0,
            'fobt': 0.1,        # ì €ë ´
            'colonoscopy': 0.5  # ë¹„ìŒˆ (ê¸°ì¤€ì )
        }
        self.reward = {
            'alive': 0.1,       # ìƒì¡´ ë³´ìƒ
            'polyp_removed': 5.0, # ì˜ˆë°© ë³´ìƒ
            'cancer_detected': 100.0, # ì¡°ê¸° ë°œê²¬ ë³´ìƒ
            'death_penalty': -5000.0 # ì‚¬ë§ í˜ë„í‹°
        }

    def _interpolate_risk(self, percentiles):
        """
        Risk_Settings.m ì˜ 'AdjustRiskGraph' í•¨ìˆ˜ ë¡œì§ êµ¬í˜„
        ì£¼ì–´ì§„ ë°±ë¶„ìœ„ìˆ˜ ì ë“¤ì„ ì„ í˜• ë³´ê°„í•˜ì—¬ ì „ì²´ ì¸êµ¬ì˜ ìœ„í—˜ë„ ë¶„í¬ë¥¼ ìƒì„±
        """
        # xì¶• ê¸°ì¤€ì  (MATLAB ì½”ë“œì˜ Values ë°°ì—´)
        x_points = np.array([10, 20, 30, 40, 50, 60, 70, 80, 90, 95, 97, 100]) * 5
        
        # 0~500 ë²”ìœ„ì˜ ë¶„í¬ ìƒì„±
        risk_dist = np.zeros(501)
        risk_dist[:x_points[0]] = percentiles[0]
        
        prev_x = x_points[0]
        prev_y = percentiles[0]
        
        for i in range(1, len(x_points)):
            curr_x = x_points[i]
            curr_y = percentiles[i]
            
            # ì„ í˜• ë³´ê°„
            for x in range(prev_x, curr_x):
                risk_dist[x] = prev_y + (curr_y - prev_y) * (x - prev_x) / (curr_x - prev_x)
            
            prev_x = curr_x
            prev_y = curr_y
            
        return risk_dist
```

```python
#based on NumberCrunching_1000.m
import gymnasium as gym
import numpy as np
from gymnasium import spaces
from cmost_config import CMOSTConfig

class CmostEnv(gym.Env):
    """
    CMOST ê¸°ë°˜ ëŒ€ì¥ì•” ìŠ¤í¬ë¦¬ë‹ ê°•í™”í•™ìŠµ í™˜ê²½
    Reference: NumberCrunching_10000.m
    """
    def __init__(self):
        super(CmostEnv, self).__init__()
        self.cfg = CMOSTConfig()
        
        # --- Action Space ---
        # 0: ëŒ€ê¸° (Wait)
        # 1: ë¶„ë³€ì í˜ˆê²€ì‚¬ (FOBT)
        # 2: ëŒ€ì¥ë‚´ì‹œê²½ (Colonoscopy)
        self.action_space = spaces.Discrete(3)
        
        # --- Observation Space ---
        # [0]: í˜„ì¬ ë‚˜ì´ (normalized 0~1)
        # [1]: ì¦ìƒ ìœ ë¬´ (0 or 1) - Symptom develops logic
        # [2]: ë§ˆì§€ë§‰ ê²€ì‚¬ ì¢…ë¥˜ (0:ì—†ìŒ, 1:FOBT, 2:Colo)
        # [3]: ë§ˆì§€ë§‰ ê²€ì‚¬ ê²°ê³¼ (0:ì •ìƒ, 1:ìš©ì¢…, 2:ì•”)
        # [4]: ë§ˆì§€ë§‰ ê²€ì‚¬ë¡œë¶€í„° ê²½ê³¼ ì‹œê°„ (normalized)
        self.observation_space = spaces.Box(low=0, high=1, shape=(5,), dtype=np.float32)
        
        self.reset()

    def reset(self, seed=None, options=None):
        super().reset(seed=seed)
        
        # 1. í™˜ì ì´ˆê¸°í™” (NumberCrunching.m: INITIALIZE ë¶€ë¶„)
        self.age = 50 # ì‹œì‘ ë‚˜ì´
        self.max_age = 90
        
        # í™˜ì ê³ ìœ  ìœ„í—˜ë„ í• ë‹¹ (Risk_Settings.m: IndividualRisk ì ìš©)
        self.individual_risk = np.random.choice(self.cfg.individual_risk_dist)
        
        # Hidden States (Preclinical - AgentëŠ” ë³¼ ìˆ˜ ì—†ìŒ)
        # ìš©ì¢… ë¦¬ìŠ¤íŠ¸: ê° ìš”ì†ŒëŠ” ìš©ì¢…ì˜ í¬ê¸° ë‹¨ê³„ (1~6)
        self.hidden_polyps = [] 
        # ì•” ìƒíƒœ: None ë˜ëŠ” ë³‘ê¸° (1~4)
        self.hidden_cancer = None 
        # ì¦ìƒ íƒ€ì´ë¨¸ (ì•” ë°œìƒ ì‹œ ì„¤ì •ë¨)
        self.symptom_timer = -1 
        
        # Observation History
        self.last_test_type = 0
        self.last_test_result = 0
        self.time_since_last_test = 0
        
        return self._get_obs(), {}

    def step(self, action):
        reward = 0
        done = False
        info = {}
        
        # ====================================================
        # 1. Action ìˆ˜í–‰ (Screening_Settings.m & Colonoscopy í•¨ìˆ˜)
        # ====================================================
        test_performed = False
        
        if action == 1: # FOBT
            reward -= self.cfg.cost['fobt']
            test_performed = True
            detected, finding_type = self._perform_screening('fobt')
            
            if detected:
                self.last_test_result = finding_type # 1:ìš©ì¢…, 2:ì•”
                # FOBT ì–‘ì„±ì´ë©´ ë³´í†µ ë‚´ì‹œê²½ì„ í•˜ëŸ¬ ê°€ë¯€ë¡œ(ì¹˜ë£Œ ê¸°íšŒ) ì•½ê°„ì˜ ë³´ìƒ
                reward += 0.5 
            else:
                self.last_test_result = 0

        elif action == 2: # Colonoscopy
            reward -= self.cfg.cost['colonoscopy']
            test_performed = True
            detected, finding_type = self._perform_screening('colonoscopy', remove_polyp=True)
            
            if detected:
                self.last_test_result = finding_type
                if finding_type == 1: # ìš©ì¢… ë°œê²¬ ë° ì œê±°
                    reward += self.cfg.reward['polyp_removed']
                elif finding_type == 2: # ì•” ë°œê²¬
                    reward += self.cfg.reward['cancer_detected']
                    # ì•” ë°œê²¬ ì‹œ ì‹œë®¬ë ˆì´ì…˜ ì¢…ë£Œ (ì¹˜ë£Œ ëª¨ë“œë¡œ ì „í™˜ ê°€ì •)
                    done = True 
            else:
                self.last_test_result = 0

        if test_performed:
            self.last_test_type = action
            self.time_since_last_test = 0
        else:
            self.time_since_last_test += 1

        # ====================================================
        # 2. ìì—°ì‚¬ ì§„í–‰ (NumberCrunching.m: Natural History)
        # ====================================================
        self._update_natural_history()

        # ====================================================
        # 3. ì¢…ë£Œ ì¡°ê±´ (Death / Max Age)
        # ====================================================
        self.age += 1
        reward += self.cfg.reward['alive'] # 1ë…„ ìƒì¡´ ë³´ìƒ
        
        # (1) ì•”ìœ¼ë¡œ ì¸í•œ ì‚¬ë§ ì²´í¬ (ì¦ìƒ ë°œí˜„ í›„ ë°©ì¹˜ëœ ê²½ìš°)
        if self.hidden_cancer and self.symptom_timer == 0:
            # ì¦ìƒì´ ë°œí˜„ë˜ì—ˆëŠ”ë°(Symptom develops), ì¹˜ë£Œë˜ì§€ ì•ŠìŒ -> ì‚¬ë§ ìœ„í—˜
            if np.random.rand() < 0.5: # ê°„ë‹¨í™”ëœ ì‚¬ë§ í™•ë¥ 
                done = True
                reward += self.cfg.reward['death_penalty']
                info['cause_of_death'] = 'cancer'

        # (2) ìì—°ì‚¬ (LifeTable ì ìš© ê°€ëŠ¥í•˜ë‚˜ ì—¬ê¸°ì„  90ì„¸ë¡œ ë‹¨ìˆœí™”)
        if self.age >= self.max_age:
            done = True
            info['cause_of_death'] = 'natural'

        return self._get_obs(), reward, done, False, info

    def _update_natural_history(self):
        """ NumberCrunching.m ì˜ í•µì‹¬ í™•ë¥  ë¡œì§ ì´ì‹ """
        
        # A. ìš©ì¢… ë°œìƒ (New Polyp)
        # MATLAB: if rand < PolypRate(z) ...
        current_onset_prob = self.cfg.prob_onset_base * self.individual_risk
        if np.random.rand() < current_onset_prob:
            # í¬ê¸° 1ë‹¨ê³„ ìš©ì¢… ìƒì„±
            self.hidden_polyps.append(1) 

        # B. ìš©ì¢… ì„±ì¥ (Progression)
        # MATLAB: AgeProgression * LocationProgression ...
        for i in range(len(self.hidden_polyps)):
            # ê°„ë‹¨í™”: ì¼ì • í™•ë¥ ë¡œ í¬ê¸° +1
            if np.random.rand() < self.cfg.prob_growth:
                self.hidden_polyps[i] += 1
        
        # C. ì•”í™” (Malignancy)
        # MATLAB: if Polyp.Polyps(z, f) > 6 ... Ca.Cancer(z, l2) = 7
        # ìš©ì¢…ì´ 6ë‹¨ê³„ë¥¼ ë„˜ì–´ê°€ë©´ ì•”ìœ¼ë¡œ ë³€í•¨
        for i in reversed(range(len(self.hidden_polyps))):
            if self.hidden_polyps[i] > 6:
                if self.hidden_cancer is None:
                    self.hidden_cancer = 1 # 1ê¸° ì•” ë°œìƒ
                    self.symptom_timer = 5 # 5ë…„ ë’¤ ì¦ìƒ ë°œí˜„ (Sojourn time)
                self.hidden_polyps.pop(i) # ìš©ì¢…ì€ ì•”ì´ ë˜ì—ˆìœ¼ë¯€ë¡œ ë¦¬ìŠ¤íŠ¸ì—ì„œ ì œê±°

        # D. ì•” ì§„í–‰ (Cancer Progression)
        if self.hidden_cancer is not None:
            if self.hidden_cancer < 4: # 4ê¸°ê¹Œì§€ ì§„í–‰
                if np.random.rand() < self.cfg.prob_cancer_progress:
                    self.hidden_cancer += 1
            
            # ì¦ìƒ íƒ€ì´ë¨¸ ê°ì†Œ
            if self.symptom_timer > 0:
                self.symptom_timer -= 1

    def _perform_screening(self, test_type, remove_polyp=False):
        """ Screening_Settings.m ì˜ ë¯¼ê°ë„ ì ìš© """
        sens_table = self.cfg.sensitivities[test_type]
        detected = False
        finding_type = 0 # 0:ì—†ìŒ, 1:ìš©ì¢…, 2:ì•”
        
        # 1. ì•” íƒì§€ ì‹œë„
        if self.hidden_cancer is not None:
            # ì•” ë³‘ê¸°(1~4)ì— ë”°ë¥¸ ë¯¼ê°ë„ ì¸ë±ìŠ¤: 6~9 (Python index 6,7,8,9)
            sens_idx = 5 + self.hidden_cancer 
            sensitivity = sens_table[sens_idx]
            if np.random.rand() < sensitivity:
                detected = True
                finding_type = 2
                return detected, finding_type # ì•” ì°¾ìœ¼ë©´ ì¦‰ì‹œ ë¦¬í„´

        # 2. ìš©ì¢… íƒì§€ ì‹œë„
        # ë°œê²¬ëœ ìš©ì¢…ë“¤ì„ ì €ì¥í•  ì„ì‹œ ë¦¬ìŠ¤íŠ¸
        detected_polyps_indices = []
        
        for i, size in enumerate(self.hidden_polyps):
            # ìš©ì¢… í¬ê¸°(1~6)ì— ë”°ë¥¸ ë¯¼ê°ë„ ì¸ë±ìŠ¤: 0~5
            sens_idx = min(size - 1, 5)
            sensitivity = sens_table[sens_idx]
            
            if np.random.rand() < sensitivity:
                detected = True
                finding_type = 1
                detected_polyps_indices.append(i)
        
        # ëŒ€ì¥ë‚´ì‹œê²½ì¸ ê²½ìš° ë°œê²¬ëœ ìš©ì¢… ì œê±° (Polypectomy)
        if remove_polyp and detected_polyps_indices:
            # ì¸ë±ìŠ¤ê°€ ê¼¬ì´ì§€ ì•Šê²Œ ë’¤ì—ì„œë¶€í„° ì œê±°
            for i in sorted(detected_polyps_indices, reverse=True):
                self.hidden_polyps.pop(i)
                
        return detected, finding_type

    def _get_obs(self):
        # Normalize age
        norm_age = (self.age - 50) / (90 - 50)
        
        # ì¦ìƒ ìœ ë¬´ (ì•”ì´ ìˆê³  íƒ€ì´ë¨¸ê°€ 0ì´ë©´ ì¦ìƒ ìˆìŒ)
        has_symptom = 1.0 if (self.hidden_cancer and self.symptom_timer <= 0) else 0.0
        
        # ê²½ê³¼ ì‹œê°„ Normalize (ìµœëŒ€ 10ë…„ìœ¼ë¡œ ê°€ì •)
        norm_time = min(self.time_since_last_test, 10) / 10.0
        
        return np.array([
            norm_age,
            has_symptom,
            self.last_test_type / 2.0,    # 0, 0.5, 1.0
            self.last_test_result / 2.0,  # 0, 0.5, 1.0
            norm_time
        ], dtype=np.float32)
```

```python
#RL training script
#vectorization of env -> PPO model creation -> Learn
import gymnasium as gym
import numpy as np
from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3.common.evaluation import evaluate_policy
from stable_baselines3.common.callbacks import EvalCallback

# --- [0] ì´ì „ì— ë§Œë“  Configì™€ Env í´ë˜ìŠ¤ê°€ ìœ„ì— ì •ì˜ë˜ì–´ ìˆë‹¤ê³  ê°€ì •í•©ë‹ˆë‹¤ ---
# (ë§Œì•½ ì •ì˜ ì•ˆ ë˜ì–´ ìˆë‹¤ë©´, ì´ì „ ë‹µë³€ì˜ ì½”ë“œë¥¼ ë¨¼ì € ì‹¤í–‰í•´ì£¼ì„¸ìš”)

# --- [1] í™˜ê²½ ìƒì„± ë° ë˜í•‘ ---
# DummyVecEnvëŠ” ì—¬ëŸ¬ í™˜ê²½ì„ ë™ì‹œì— ëŒë¦´ ìˆ˜ ìˆê²Œ í•´ì£¼ëŠ”ë°, 
# ì—¬ê¸°ì„œëŠ” 1ê°œë§Œ ëŒë¦¬ì§€ë§Œ í˜¸í™˜ì„±ì„ ìœ„í•´ ì‚¬ìš©í•©ë‹ˆë‹¤.
env = DummyVecEnv([lambda: CmostEnv()])

# --- [2] DRL ëª¨ë¸ ì •ì˜ (PPO ì•Œê³ ë¦¬ì¦˜) ---
# MlpPolicy: ì…ë ¥ì´ ì´ë¯¸ì§€(CNN)ê°€ ì•„ë‹ˆë¼ ìˆ˜ì¹˜ ë²¡í„°(Vector)ì¼ ë•Œ ì‚¬ìš©í•˜ëŠ” ì‹ ê²½ë§ êµ¬ì¡°
# verbose=1: í•™ìŠµ ê³¼ì •ì„ ë¡œê·¸ë¡œ ì¶œë ¥
model = PPO(
    "MlpPolicy", 
    env, 
    verbose=1, 
    learning_rate=0.0003,    # í•™ìŠµë¥  (ë„ˆë¬´ í¬ë©´ ë°œì‚°, ë„ˆë¬´ ì‘ìœ¼ë©´ ëŠë¦¼)
    gamma=0.999,              # í• ì¸ìœ¨ (ë¯¸ë˜ ë³´ìƒì„ ì–¼ë§ˆë‚˜ ì¤‘ìš”í•˜ê²Œ ì—¬ê¸¸ì§€)
    ent_coef=0.01            # íƒí—˜ ê³„ìˆ˜ (ìƒˆë¡œìš´ í–‰ë™ì„ ì‹œë„í•´ë³´ëŠ” ë¹„ìœ¨)
)

print("ğŸš€ í•™ìŠµ ì‹œì‘! (ì ì‹œë§Œ ê¸°ë‹¤ë ¤ì£¼ì„¸ìš”...)")

# --- [3] ëª¨ë¸ í•™ìŠµ (Training) ---
# total_timesteps: ì—ì´ì „íŠ¸ê°€ í™˜ê²½ê³¼ ìƒí˜¸ì‘ìš©í•˜ëŠ” ì´ íšŸìˆ˜ (ë‹¨ìœ„: ë…„)
# 10ë§Œ ë²ˆ ì •ë„ë©´ ê°„ë‹¨í•œ ì •ì±…ì€ ë°°ìš¸ ìˆ˜ ìˆìŠµë‹ˆë‹¤. (ì„±ëŠ¥ ë†’ì´ë ¤ë©´ 100ë§Œ ì´ìƒ ì¶”ì²œ)
model.learn(total_timesteps=500000)

print("âœ… í•™ìŠµ ì™„ë£Œ!")

# --- [4] ëª¨ë¸ ì €ì¥ ---
model.save("cmost_ppo_agent")
print("ğŸ’¾ ëª¨ë¸ì´ 'cmost_ppo_agent.zip'ìœ¼ë¡œ ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")
```

```python
#evaluation (vs random)

# --- [5] í•™ìŠµëœ ëª¨ë¸ ì„±ëŠ¥ í…ŒìŠ¤íŠ¸ ---

def run_simulation(agent_model, episodes=10):
    """ ì—ì´ì „íŠ¸ë¥¼ ì´ìš©í•´ í™˜ì ì‹œë®¬ë ˆì´ì…˜ì„ ëŒë¦¬ê³  ê²°ê³¼ë¥¼ ì¶œë ¥í•˜ëŠ” í•¨ìˆ˜ """
    # í…ŒìŠ¤íŠ¸ìš© í™˜ê²½ ë³„ë„ ìƒì„±
    test_env = CmostEnv()
    total_rewards = []
    lifespans = []
    deaths_by_cancer = 0
    
    print(f"\n--- ì‹œë®¬ë ˆì´ì…˜ ì‹œì‘ ({episodes}ëª… í™˜ì) ---")
    
    for ep in range(episodes):
        obs, _ = test_env.reset()
        done = False
        ep_reward = 0
        actions = []
        
        while not done:
            # ëª¨ë¸ì´ ìˆìœ¼ë©´ ì˜ˆì¸¡ê°’ ì‚¬ìš©, ì—†ìœ¼ë©´ ëœë¤ í–‰ë™
            if agent_model:
                action, _ = agent_model.predict(obs, deterministic=True)
            else:
                action = test_env.action_space.sample()
                
            obs, reward, done, _, info = test_env.step(action)
            ep_reward += reward
            actions.append(action)
            
        total_rewards.append(ep_reward)
        lifespans.append(test_env.age)
        if info.get('cause_of_death') == 'cancer':
            deaths_by_cancer += 1
            
        # ì²˜ìŒ 3ëª…ë§Œ ë¡œê·¸ ìì„¸íˆ ì¶œë ¥
        if ep < 3:
            print(f"í™˜ì {ep+1}: ì‚¬ë§ë‚˜ì´ {test_env.age}ì„¸, ì‚¬ì¸: {info.get('cause_of_death')}")
            # í–‰ë™ í†µê³„ (0:ëŒ€ê¸°, 1:FOBT, 2:ë‚´ì‹œê²½)
            unique, counts = np.unique(actions, return_counts=True)
            action_dict = dict(zip(unique, counts))
            print(f"   -> ê²€ì‚¬ íšŸìˆ˜: ë‚´ì‹œê²½ {action_dict.get(2, 0)}íšŒ, FOBT {action_dict.get(1, 0)}íšŒ")
            print(f"   -> ì´ ë³´ìƒ: {ep_reward:.2f}")

    print("-" * 30)
    print(f"í‰ê·  ìˆ˜ëª…: {np.mean(lifespans):.1f}ì„¸")
    print(f"í‰ê·  ë³´ìƒ: {np.mean(total_rewards):.2f}")
    print(f"ì•” ì‚¬ë§ë¥ : {(deaths_by_cancer/episodes)*100:.1f}%")

# 1. í•™ìŠµ ì „ (ëœë¤) ì„±ëŠ¥
print("\n[ë¹„êµ 1] í•™ìŠµ ì „ (Random Agent)")
run_simulation(None, episodes=100)

# 2. í•™ìŠµ í›„ (PPO) ì„±ëŠ¥
print("\n[ë¹„êµ 2] í•™ìŠµ í›„ (PPO Agent)")
run_simulation(model, episodes=100)
```

```python
#config
import numpy as np
import scipy.io

class CMOSTConfig:
    def __init__(self, mat_file_path='cmost/Settings/CMOST13.mat'):
        print(f"Loading Optimized Configuration from {mat_file_path}...")
        
        try:
            mat_data = scipy.io.loadmat(mat_file_path)
            self.data = mat_data['temp'][0, 0]
            print("âœ… ë°ì´í„° ë¡œë“œ ì„±ê³µ: í•µì‹¬ ë³€ìˆ˜ ì¶”ì¶œ ë° ë³´ê°„ ì‹œì‘")
        except Exception as e:
            print(f"âŒ ë°ì´í„° ë¡œë“œ ì‹¤íŒ¨: {e}")
            self.data = None

        if self.data is not None:
            # =========================================================
            # [1] ì¸êµ¬í•™ì  í†µê³„ (Demographics)
            # =========================================================
            # ì—¬ì„± ë¹„ìœ¨ (ì„±ë³„ì— ë”°ë¥¸ ë°œë³‘ë¥  ì°¨ì´ ì ìš©ì„ ìœ„í•´ í•„ìˆ˜)
            self.frac_female = self._get_scalar('fraction_female')

            # =========================================================
            # [2] ìƒë¬¼í•™ì  íŒŒë¼ë¯¸í„° (Natural History) - ì„±ë³„ êµ¬ë¶„
            # =========================================================
            # ë‚¨ì„± ë°ì´í„° (ê¸°ë³¸ê°’) -> 100ì„¸ê¹Œì§€ ë³´ê°„(Interpolation)
            self.male_onset = self._interpolate(self.data['NewPolypRate'])
            self.male_early = self._interpolate(self.data['EarlyProgressionRate'])
            self.male_adv   = self._interpolate(self.data['AdvancedProgressionRate'])
            # LifeTable: 0ë²ˆ ì»¬ëŸ¼=ë‚¨ì„±, 1ë²ˆ ì»¬ëŸ¼=ì—¬ì„±
            self.male_life  = self.data['LifeTable'][:, 0]

            # ì—¬ì„± ë°ì´í„° (ë³€ìˆ˜ëª…ì´ ì¡´ì¬í•˜ë©´ ë¡œë“œ, ì—†ìœ¼ë©´ ë‚¨ì„± ë°ì´í„° ì‚¬ìš©)
            if 'new_polyp_female' in self.data.dtype.names:
                self.female_onset = self._interpolate(self.data['new_polyp_female'])
                self.female_early = self._interpolate(self.data['early_progression_female'])
                self.female_adv   = self._interpolate(self.data['advanced_progression_female'])
                self.female_life  = self.data['LifeTable'][:, 1]
            else:
                self.female_onset, self.female_early, self.female_adv = self.male_onset, self.male_early, self.male_adv
                self.female_life = self.male_life

            # [ê³µí†µ ìƒë¬¼í•™]
            # Healing: ìš©ì¢…ì´ ì €ì ˆë¡œ ì‚¬ë¼ì§ˆ í™•ë¥  (í¬ê¸°ë³„)
            self.prob_healing = self.data['Healing'].flatten() 
            # FastCancer: ìš©ì¢… ë‹¨ê³„ë¥¼ ê±´ë„ˆë›°ê³  ë¹ ë¥´ê²Œ ì•”ì´ ë  í™•ë¥ 
            self.prob_fast_cancer = self.data['FastCancer'].flatten()
            # Symptoms: ì•” ê¸°ìˆ˜ë³„ ì¦ìƒ ë°œí˜„ìœ¨
            self.prob_symptoms = self.data['Symptoms'].flatten()
            # IndividualRisk: ì‚¬ëŒë§ˆë‹¤ ë‹¤ë¥¸ ìœ„í—˜ë„ ë¶„í¬
            self.risk_dist = self.data['IndividualRisk'].flatten()

            # =========================================================
            # [3] ê²€ì‚¬ ì„±ëŠ¥ ë° í•©ë³‘ì¦ (Screening & Complications)
            # =========================================================
            # ë‚´ì‹œê²½ ë¯¼ê°ë„ (ìš©ì¢… í¬ê¸°ë³„ ë°œê²¬ í™•ë¥ )
            self.sens_colo = self.data['Colo_Detection'].flatten()
            
            # ë‚´ì‹œê²½ í•©ë³‘ì¦ í™•ë¥  (ì²œê³µ, ì¶œí˜ˆ) -> AIê°€ ë¬´ë¶„ë³„í•œ ê²€ì‚¬ë¥¼ ëª»í•˜ê²Œ ë§‰ëŠ” ì œë™ì¥ì¹˜
            self.risk_perf  = self._get_scalar('Colonoscopy_RiscPerforation')
            self.risk_bleed = self._get_scalar('Colonoscopy_RiscBleeding')
            self.death_perf = self._get_scalar('DeathPerforation') # ì²œê³µ ì‹œ ì‚¬ë§ë¥ 

        else:
            # ë¡œë“œ ì‹¤íŒ¨ ì‹œ ê¸°ë³¸ ë”ë¯¸ ê°’ (ì˜¤ë¥˜ ë°©ì§€ìš©)
            self._set_dummy_values()

        # [4] ê³ ì •ê°’ (Benchmarkì—ì„œ ê°€ì ¸ì˜¨ FOBT ì„±ëŠ¥)
        self.sens_fobt = np.array([0.02, 0.02, 0.05, 0.05, 0.12, 0.12, 0.4, 0.4, 0.4, 0.4])

        # [5] ê°•í™”í•™ìŠµ ë³´ìƒ ì²´ê³„ (Reward Function)
        self.rewards = {
            'cost_wait': 0,
            'cost_fobt': 0.1,
            'cost_colo': 0.5,           # ë‚´ì‹œê²½ ë¹„ìš©
            'reward_alive': 0.1,        # 1ë…„ ìƒì¡´ ë³´ìƒ
            'reward_polyp': 5.0,        # ìš©ì¢… ì œê±° (ì˜ˆë°©)
            'reward_cancer': 100.0,     # ì•” ë°œê²¬ (ì¹˜ë£Œ)
            'penalty_death': -5000.0,   # ì‚¬ë§ í˜ë„í‹°
            'penalty_complication': -50.0 # í•©ë³‘ì¦(ì²œê³µ/ì¶œí˜ˆ) ë°œìƒ ì‹œ í˜ë„í‹°
        }

    # --- ë‚´ë¶€ í—¬í¼ í•¨ìˆ˜ë“¤ ---
    def _get_scalar(self, key):
        """ ë³€ìˆ˜ê°€ ìˆìœ¼ë©´ ê°’(float)ì„ êº¼ë‚´ê³ , ì—†ìœ¼ë©´ 0.0 ë°˜í™˜ """
        if key in self.data.dtype.names:
            val = self.data[key]
            return val.item() if val.size == 1 else val.flatten()[0]
        return 0.0

    def _interpolate(self, raw_data, target_len=100):
        """ (1, 20) ë°ì´í„°ë¥¼ 100ì„¸ê¹Œì§€ ì„ í˜• ë³´ê°„ """
        if raw_data is None or raw_data.size == 0: return np.zeros(target_len)
        y = raw_data.flatten()
        x = np.linspace(0, target_len, len(y))
        return np.interp(np.arange(target_len), x, y)
    
    def _set_dummy_values(self):
        """ íŒŒì¼ ì—†ì„ ë•Œë¥¼ ìœ„í•œ ê¸°ë³¸ê°’ """
        self.male_onset = np.linspace(0.01, 0.05, 100)
        self.male_early = np.full(100, 0.1)
        self.male_adv = np.full(100, 0.1)
        self.male_life = np.linspace(0, 1, 100)
        self.female_onset = self.male_onset
        self.prob_healing = np.zeros(10)
        self.risk_dist = np.ones(100)
        self.sens_colo = np.full(10, 0.9)
        self.risk_perf = 0.001

    # --- ì™¸ë¶€ í˜¸ì¶œìš© ì¸í„°í˜ì´ìŠ¤ ---
    def get_rates(self, is_female):
        """ ì„±ë³„ì— ë§ëŠ” í™•ë¥  ì„¸íŠ¸(ë°œìƒ, ì„±ì¥, ì•”í™”, ì‚¬ë§) ë°˜í™˜ """
        if is_female:
            return self.female_onset, self.female_early, self.female_adv, self.female_life
        return self.male_onset, self.male_early, self.male_adv, self.male_life
    
    def get_sensitivity(self, test_type, state_idx):
        """ ê²€ì‚¬ ì¢…ë¥˜ì™€ ë³‘ë³€ ìƒíƒœ(0~9)ì— ë”°ë¥¸ ë¯¼ê°ë„ ë°˜í™˜ """
        if state_idx < 0: return 0
        idx = min(state_idx, 9) # ì¸ë±ìŠ¤ ì´ˆê³¼ ë°©ì§€
        
        if test_type == 'colonoscopy':
            return self.sens_colo[idx]
        elif test_type == 'fobt':
            return self.sens_fobt[idx]
        return 0
```

```python
#env
import gymnasium as gym
from gymnasium import spaces
import numpy as np

class CmostEnv(gym.Env):
    def __init__(self):
        super(CmostEnv, self).__init__()
        self.cfg = CMOSTConfig() # ì•ì„œ ë§Œë“  Config (Risk, Gender í¬í•¨ëœ ë²„ì „)
        
        # [Action Space] 0:Wait, 1:FOBT, 2:Colonoscopy
        self.action_space = spaces.Discrete(3)
        
        # [Observation Space]
        # 0: ë‚˜ì´, 1: ì„±ë³„, 2: ì¦ìƒ, 3: ë§ˆì§€ë§‰ê²€ì‚¬, 4: ê²°ê³¼, 5: ê²½ê³¼ì‹œê°„
        self.observation_space = spaces.Box(low=0, high=1, shape=(6,), dtype=np.float32)

    # def reset(self, seed=None, options=None):
    #     super().reset(seed=seed)
        
    #     # -------------------------------------------------------
    #     # 1. í™˜ì ìƒì„± (Initialization)
    #     # -------------------------------------------------------
    #     self.max_age = 100 # MATLAB: y < 100
        
    #     # ì„±ë³„ ë° ê°œì¸ ìœ„í—˜ë„ ê²°ì • (MATLAB Initialize íŒŒíŠ¸)
    #     self.is_female = np.random.rand() < self.cfg.frac_female
    #     self.risk_factor = np.random.choice(self.cfg.risk_dist)
        
    #     # ì„±ë³„ì— ë”°ë¥¸ í™•ë¥ í‘œ ì„¸íŒ…
    #     self.rates = self.cfg.get_rates(self.is_female) 
    #     # (onset, early, adv, life_table) íŠœí”Œ ì–¸íŒ¨í‚¹
    #     self.r_onset, self.r_early, self.r_adv, self.r_life = self.rates

    #     # ìƒíƒœ ë³€ìˆ˜ ì´ˆê¸°í™”
    #     self.hidden_polyps = []   # ìš©ì¢… í¬ê¸° ë¦¬ìŠ¤íŠ¸ (1~9ë‹¨ê³„)
    #     self.hidden_cancer = None # ì•” ê¸°ìˆ˜ (None, 1, 2, 3, 4)
        
    #     # -------------------------------------------------------
    #     # 2. Warm-up Simulation (0ì„¸ -> 50ì„¸)
    #     # MATLABì€ ì²˜ìŒë¶€í„° ëŒë¦¬ë¯€ë¡œ, ìš°ë¦¬ë„ 50ì„¸ ìƒíƒœë¥¼ ë§Œë“¤ì–´ì•¼ í•¨
    #     # -------------------------------------------------------
    #     self.age = 0
    #     target_start_age = 50
        
    #     # 50ì„¸ê°€ ë  ë•Œê¹Œì§€ ì¹˜ë£Œ ì—†ì´ ìì—°ì‚¬(Biology)ë§Œ ì§„í–‰
    #     for _ in range(target_start_age):
    #         # ë§¤ë…„ 4ë¶„ê¸°(Quarterly) ì—…ë°ì´íŠ¸
    #         for q in range(4):
    #             self._update_biology_quarterly(self.age + q/4.0)
    #         self.age += 1
            
    #     # 50ì„¸ ì‹œì ì˜ ìƒíƒœ ì„¤ì •
    #     self.last_test_type = 0
    #     self.last_test_result = 0
    #     self.time_since_test = 10.0 # 10ë…„ ì´ìƒ (ê²€ì‚¬ ì•ˆ í•¨)
    #     self.has_symptom = 0
        
    #     # ë§Œì•½ Warm-up ë„ì¤‘ ì•”ìœ¼ë¡œ ì‚¬ë§í–ˆê±°ë‚˜ ì§„í–‰ëë‹¤ë©´ ë‹¤ì‹œ ë¦¬ì…‹ (RL í•™ìŠµ íš¨ìœ¨ ìœ„í•´)
    #     if self.hidden_cancer and self.hidden_cancer >= 4:
    #         return self.reset() # ë„ˆë¬´ ì•„í”ˆ í™˜ìëŠ” ìŠ¤í‚µí•˜ê³  ë‹¤ì‹œ ë½‘ê¸°

    #     return self._get_obs(), {}

    def reset(self, seed=None, options=None):
        super().reset(seed=seed)
        
        # [ìˆ˜ì •] ë¬´í•œ ì¬ê·€ ë°©ì§€ë¥¼ ìœ„í•œ ë£¨í”„ ë°©ì‹ ë„ì…
        max_retries = 100 # ìµœëŒ€ 100ë²ˆê¹Œì§€ë§Œ ë‹¤ì‹œ ì‹œë„
        attempt = 0
        
        while attempt < max_retries:
            attempt += 1
            
            # -------------------------------------------------------
            # 1. í™˜ì ìƒì„± (Initialization)
            # -------------------------------------------------------
            self.max_age = 100
            
            # ì„±ë³„ ë° ê°œì¸ ìœ„í—˜ë„ ê²°ì •
            self.is_female = np.random.rand() < self.cfg.frac_female
            self.risk_factor = np.random.choice(self.cfg.risk_dist)
            
            # ì„±ë³„ì— ë”°ë¥¸ í™•ë¥ í‘œ ì„¸íŒ…
            self.rates = self.cfg.get_rates(self.is_female) 
            self.r_onset, self.r_early, self.r_adv, self.r_life = self.rates

            # ìƒíƒœ ë³€ìˆ˜ ì´ˆê¸°í™”
            self.hidden_polyps = []   
            self.hidden_cancer = None 
            
            # -------------------------------------------------------
            # 2. Warm-up Simulation (0ì„¸ -> 50ì„¸)
            # -------------------------------------------------------
            self.age = 0
            target_start_age = 50
            is_dead_during_warmup = False
            
            # 50ì„¸ê°€ ë  ë•Œê¹Œì§€ ìì—°ì‚¬ ì§„í–‰
            for _ in range(target_start_age):
                for q in range(4):
                    current_time = self.age + q/4.0
                    self._update_biology_quarterly(current_time)
                    
                    # ë§Œì•½ Warm-up ë„ì¤‘ ìì—°ì‚¬í–ˆê±°ë‚˜ ì•”ìœ¼ë¡œ ë„ˆë¬´ ì•…í™”ë˜ë©´ ì‹¤íŒ¨ ì²˜ë¦¬
                    if self._check_death(current_time): # ìì—°ì‚¬ ì²´í¬
                        is_dead_during_warmup = True
                        break
                        
                if is_dead_during_warmup: break
                self.age += 1
                
            # [ì¡°ê±´ ì²´í¬] 
            # 1. Warm-up ì¤‘ ì£½ì§€ ì•Šì•˜ì–´ì•¼ í•¨
            # 2. 50ì„¸ ì‹œì ì— ì´ë¯¸ ë§ê¸° ì•”(4ê¸°)ì´ë©´ ì•ˆ ë¨ (RL í•™ìŠµìš© í™˜ìë¡œ ë¶€ì í•©)
            if not is_dead_during_warmup:
                if not (self.hidden_cancer and self.hidden_cancer >= 4):
                    break # ì„±ê³µ! ë£¨í”„ íƒˆì¶œ
        
        # -------------------------------------------------------
        # [ì•ˆì „ì¥ì¹˜] 100ë²ˆ ì‹œë„í•´ë„ ì‹¤íŒ¨í•˜ë©´? -> ê°•ì œë¡œ ì´ˆê¸°í™” (Reset Force)
        # -------------------------------------------------------
        if attempt >= max_retries:
            # "ìš´ ë‚˜ìœ í™˜ìë“¤ì€ ë‹¤ ê±¸ëŸ¬ë‚´ê³ , ê·¸ëƒ¥ í‰ë²”í•œ í™˜ìë¡œ ê°•ì œ ì‹œì‘"
            self.age = 50
            self.hidden_polyps = []
            self.hidden_cancer = None # ì•” ì—†ìŒìœ¼ë¡œ ê°•ì œ ì„¤ì •
        
        # 3. 50ì„¸ ì‹œì ì˜ ê´€ì¸¡ ì •ë³´ ì„¤ì •
        self.last_test_type = 0
        self.last_test_result = 0
        self.time_since_test = 10.0 
        self.has_symptom = 0
        
        # ì¦ìƒ ì—…ë°ì´íŠ¸ (í˜„ì¬ ì•” ìƒíƒœì— ë§ì¶°ì„œ)
        if self.hidden_cancer:
             symp_prob = self.cfg.prob_symptoms[min(5+self.hidden_cancer, 9)]
             if np.random.rand() < symp_prob:
                 self.has_symptom = 1

        return self._get_obs(), {}
    

    def step(self, action):
        reward = 0
        done = False
        info = {'cause': 'alive', 'action': action}
        
        # =========================================================
        # [Time Loop] 1 Step = 1 Year (4 Quarters)
        # MATLAB: while loop ë‚´ë¶€
        # =========================================================
        
        # 1. Action ìˆ˜í–‰ (ì—°ì´ˆì— ê²°ì •í•œë‹¤ê³  ê°€ì •)
        if not done:
            reward_act, done_act, info_act = self._perform_action(action)
            reward += reward_act
            if done_act: 
                done = True
                info.update(info_act)

        # 2. ë¶„ê¸°ë³„ ìƒë¬¼í•™ì  ë³€í™” (Quarterly Loop)
        # MATLAB: for q = 1:4
        if not done:
            for q in range(4):
                current_time = self.age + q/4.0
                
                # A. ìì—°ì‚¬ (Biology Update)
                self._update_biology_quarterly(current_time)
                
                # B. ì¦ìƒ ë°œí˜„ (Symptom Development)
                # MATLAB: if time >= Ca.SympTime
                if self.hidden_cancer:
                    symp_prob = self.cfg.prob_symptoms[min(5+self.hidden_cancer, 9)] / 4.0 # ë¶„ê¸° ë³´ì •
                    if np.random.rand() < symp_prob:
                        self.has_symptom = 1
                        # ì¦ìƒì´ ì‹¬í•˜ë©´ ë³‘ì› ë°©ë¬¸ -> ì•” ë°œê²¬ ë° ì¹˜ë£Œ ì‹œë„ (ê°„ì†Œí™”)
                        # ì—¬ê¸°ì„œëŠ” RL ì—ì´ì „íŠ¸ê°€ ì¦ìƒì„ ë³´ê³  ë‹¤ìŒ Stepì— Actionì„ ì·¨í•˜ë„ë¡ ìœ ë„
                
                # C. ì‚¬ë§ ì²´í¬ (Natural & Cancer)
                # MATLAB: people die of natural causes / cancer
                if self._check_death(current_time):
                    done = True
                    # ì•” ì‚¬ë§ì¸ì§€ ìì—°ì‚¬ì¸ì§€ íŒë³„
                    if self.hidden_cancer and self.has_symptom:
                        reward += self.cfg.rewards['penalty_death']
                        info['cause'] = 'cancer_death'
                    else:
                        info['cause'] = 'natural_death' # ìì—°ì‚¬ëŠ” í˜ë„í‹° ì—†ìŒ
                    break

        # 3. ë…„ë„ ë§ˆë¬´ë¦¬
        if not done:
            self.age += 1
            reward += self.cfg.rewards['reward_alive'] # ìƒì¡´ ë³´ìƒ
            self.time_since_test += 1
            
            if self.age >= self.max_age:
                done = True
                info['cause'] = 'max_age_reached'

        return self._get_obs(), reward, done, False, info

    # -------------------------------------------------------------
    # Helper Methods (MATLAB Logic Implementation)
    # -------------------------------------------------------------

    def _perform_action(self, action):
        """ ê²€ì‚¬ ìˆ˜í–‰ ë° ê²°ê³¼ ì²˜ë¦¬ """
        r, d = 0, False
        i = {}
        
        if action == 0: # Wait
            return r, d, i
            
        elif action == 1: # FOBT
            r -= self.cfg.rewards['cost_fobt']
            self.time_since_test = 0
            self.last_test_type = 0.5
            
            detected, f_type = self._run_test('fobt', remove=False)
            if detected:
                self.last_test_result = f_type
                r += 0.5 # ë°œê²¬ ë³´ìƒ
            else:
                self.last_test_result = 0
                
        elif action == 2: # Colonoscopy
            r -= self.cfg.rewards['cost_colo']
            self.time_since_test = 0
            self.last_test_type = 1.0
            
            # í•©ë³‘ì¦ ì²´í¬ (MATLAB: Complications if rand < risc...)
            if np.random.rand() < self.cfg.risk_perf:
                r += self.cfg.rewards['penalty_complication']
                if np.random.rand() < self.cfg.death_perf:
                    return r + self.cfg.rewards['penalty_death'], True, {'cause': 'perforation_death'}

            # ê²€ì‚¬ ìˆ˜í–‰
            detected, f_type = self._run_test('colonoscopy', remove=True)
            if detected:
                self.last_test_result = f_type
                if f_type == 1: r += self.cfg.rewards['reward_polyp']
                elif f_type == 2:
                    r += self.cfg.rewards['reward_cancer']
                    d = True # ì¹˜ë£Œ ì„±ê³µ ì¢…ë£Œ
                    i['cause'] = 'cancer_detected'
            else:
                self.last_test_result = 0
                
        return r, d, i

    def _run_test(self, t_type, remove):
        """ ë¯¼ê°ë„ ì ìš© ê²€ì‚¬ """
        det, f_type = False, 0
        
        # ì•” íƒì§€
        if self.hidden_cancer:
            sens = self.cfg.get_sensitivity(t_type, 5 + self.hidden_cancer)
            if np.random.rand() < sens: return True, 2
            
        # ìš©ì¢… íƒì§€
        rem_idx = []
        for idx, size in enumerate(self.hidden_polyps):
            sens = self.cfg.get_sensitivity(t_type, min(size-1, 5))
            if np.random.rand() < sens:
                det = True
                f_type = 1
                rem_idx.append(idx)
        
        # ë‚´ì‹œê²½ì´ë©´ ì œê±° (MATLAB: we delete the current polyp)
        if remove and rem_idx:
            for idx in sorted(rem_idx, reverse=True):
                self.hidden_polyps.pop(idx)
                
        return det, f_type

    def _update_biology_quarterly(self, current_time):
        """ 
        MATLAB NumberCrunchingì˜ í•µì‹¬ ë¡œì§ 
        (ë¶„ê¸°ë³„ë¡œ í˜¸ì¶œë˜ë¯€ë¡œ í™•ë¥ ì„ 1/4ë¡œ ì¡°ì •í•˜ì—¬ ì ìš©)
        """
        age_idx = min(int(self.age), 99)
        dt = 0.25 # 1ë¶„ê¸°
        
        # 1. New Polyp (ìš©ì¢… ë°œìƒ)
        # MATLAB: PolypRate = IndividualRisk * NewPolyp(y) * gender_factor
        # ë¶„ê¸°ë³„ í™•ë¥  ê·¼ì‚¬: 1 - (1-p)^(1/4) ~ p/4
        prob_onset = (self.r_onset[age_idx] * self.risk_factor) * dt
        if np.random.rand() < prob_onset:
            self.hidden_polyps.append(1) # í¬ê¸° 1 ìƒì„±

        # 2. Fast Cancer (Direct Cancer)
        # MATLAB: rand < DirectCancerRate * DirectCancerSpeed
        prob_fast = self.cfg.prob_fast_cancer[age_idx] if age_idx < len(self.cfg.prob_fast_cancer) else 0
        if np.random.rand() < prob_fast * dt:
            # ë°”ë¡œ ì•” ë°œìƒ (ìš©ì¢… ë‹¨ê³„ ê±´ë„ˆëœ€)
            self.hidden_cancer = 1 # 1ê¸° ì•”
            
        # 3. Polyp Progression (ìš©ì¢… ì„±ì¥)
        # MATLAB: AgeProgression * Location * Gender * Risk
        for i in range(len(self.hidden_polyps)):
            size = self.hidden_polyps[i]
            if size <= 3:
                rate = self.r_early[age_idx]
            else:
                rate = self.r_adv[age_idx]
            
            if np.random.rand() < rate * dt: # ë¶„ê¸°ë³„ ì„±ì¥ í™•ë¥ 
                self.hidden_polyps[i] += 1

        # 4. Healing (ìì—° ì¹˜ìœ /ì†Œë©¸)
        # MATLAB: a polyp shrinks or disappears
        for i in reversed(range(len(self.hidden_polyps))):
            size = self.hidden_polyps[i]
            # Healing ë°ì´í„°ê°€ 10ê°œ(ì¸ë±ìŠ¤ 0~9)ë¼ê³  ê°€ì •
            heal_prob = self.cfg.prob_healing[min(size-1, 9)] if size > 0 else 0
            if np.random.rand() < heal_prob * dt:
                self.hidden_polyps[i] -= 1 # í¬ê¸° ê°ì†Œ
                if self.hidden_polyps[i] <= 0:
                    self.hidden_polyps.pop(i) # ì†Œë©¸

        # 5. Malignancy (ì•”í™”)
        # ìš©ì¢…ì´ ë„ˆë¬´ ì»¤ì§€ë©´ ì•”ìœ¼ë¡œ ë³€í•¨
        for i in reversed(range(len(self.hidden_polyps))):
            if self.hidden_polyps[i] > 6: # 6ë‹¨ê³„ ì´ˆê³¼ ì‹œ
                # MATLABì—ì„œëŠ” ë°”ë¡œ 7(ì•”)ë¡œ ë³€í™˜
                if self.hidden_cancer is None:
                    self.hidden_cancer = 1
                self.hidden_polyps.pop(i)

        # 6. Cancer Progression (ì•” ì§„í–‰)
        if self.hidden_cancer and self.hidden_cancer < 4:
            # ì•”ì€ ë¹¨ë¦¬ ìëŒ (ì„ì˜ í™•ë¥  0.1 * dt)
            if np.random.rand() < 0.4 * dt: 
                self.hidden_cancer += 1

    def _check_death(self, current_time):
        """ ìì—°ì‚¬ ë° ì•” ì‚¬ë§ ì²´í¬ """
        age_idx = min(int(self.age), 99)
        
        # 1. ìì—°ì‚¬ (LifeTable)
        # MATLAB: if rand < (LifeTable(y)/4)
        nat_death_prob = self.r_life[age_idx] / 4.0
        if np.random.rand() < nat_death_prob:
            return True
            
        # 2. ì•” ì‚¬ë§ (Cancer Death)
        if self.hidden_cancer:
            # ì•”ì´ ì‹¬í• ìˆ˜ë¡ ì‚¬ë§ í™•ë¥  ì¦ê°€ (4ê¸°ë©´ ë§¤ìš° ë†’ìŒ)
            cancer_death_base = [0, 0.01, 0.05, 0.2, 0.8] # 0~4ê¸° ì‚¬ë§ë¥ (ì—°ê°„) ì˜ˆì‹œ
            c_death = cancer_death_base[self.hidden_cancer] / 4.0
            if np.random.rand() < c_death:
                return True
                
        return False

    def _get_obs(self):
        return np.array([
            (self.age - 50) / 50.0,      # ë‚˜ì´ (50~100 ì •ê·œí™”)
            1.0 if self.is_female else 0.0,
            self.has_symptom,
            self.last_test_type,
            self.last_test_result,
            min(self.time_since_test, 10) / 10.0
        ], dtype=np.float32)
```

```python
#ppo í•™ìŠµ. 
import gymnasium as gym
from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import DummyVecEnv
import numpy as np
import matplotlib.pyplot as plt

# ==========================================
# 1. í™˜ê²½ ì„¤ì • ë° ëª¨ë¸ ìƒì„±
# ==========================================

# DummyVecEnv: SB3ì—ì„œ í•™ìŠµ ì†ë„ë¥¼ ë†’ì´ê¸° ìœ„í•´ í™˜ê²½ì„ ê°ì‹¸ëŠ” ë˜í¼(Wrapper)
# (ì—¬ëŸ¬ í™˜ê²½ì„ ë™ì‹œì— ëŒë¦´ ìˆ˜ë„ ìˆì§€ë§Œ ì—¬ê¸°ì„  1ê°œë§Œ ì‚¬ìš©)
env = DummyVecEnv([lambda: CmostEnv()])

print("ğŸ¤– AI ëª¨ë¸ ìƒì„± ì¤‘...")

# PPO ëª¨ë¸ ì •ì˜
model = PPO(
    "MlpPolicy",         # ì…ë ¥ì´ ì´ë¯¸ì§€(CNN)ê°€ ì•„ë‹ˆë¼ ìˆ˜ì¹˜(Vector)ì´ë¯€ë¡œ MLP ì‚¬ìš©
    env, 
    verbose=1,           # í•™ìŠµ ë¡œê·¸ ì¶œë ¥ (1: ì •ë³´ ì¶œë ¥)
    learning_rate=0.0003, # í•™ìŠµ ì†ë„ (ë„ˆë¬´ ë¹ ë¥´ë©´ ë°œì‚°, ëŠë¦¬ë©´ ë‹µë‹µí•¨)
    gamma=0.999,         # [í•µì‹¬] í• ì¸ìœ¨ (0.99 -> 0.999)
                         # ë¯¸ë˜ì˜ ë³´ìƒ(ì‚¬ë§ ë°©ì§€)ì„ í˜„ì¬ì˜ ë¹„ìš©(ê²€ì‚¬ë¹„)ë³´ë‹¤ ì¤‘ìš”í•˜ê²Œ ì—¬ê¹€
    ent_coef=0.01,       # íƒí—˜ ê³„ìˆ˜ (ë‹¤ì–‘í•œ ì‹œë„ë¥¼ í•´ë³´ë„ë¡ ìœ ë„)
    batch_size=64,       # í•œ ë²ˆì— í•™ìŠµí•  ë°ì´í„° ì–‘
    seed=42              # ì¬í˜„ì„±ì„ ìœ„í•œ ì‹œë“œ ê³ ì •
)

# ==========================================
# 2. í•™ìŠµ ì „ ì„±ëŠ¥ í…ŒìŠ¤íŠ¸ (Baseline)
# ==========================================
def evaluate_agent(agent_model, num_episodes=20, name="Agent"):
    """ ì—ì´ì „íŠ¸ ì„±ëŠ¥ì„ í…ŒìŠ¤íŠ¸í•˜ëŠ” í•¨ìˆ˜ """
    test_env = CmostEnv() # í…ŒìŠ¤íŠ¸ìš© í™˜ê²½ ë³„ë„ ìƒì„±
    rewards = []
    lifespans = []
    deaths_cancer = 0
    deaths_perf = 0
    check_counts = {'fobt': 0, 'colo': 0}

    print(f"\n--- [{name}] í‰ê°€ ì‹œì‘ ({num_episodes}ëª…) ---")
    
    for ep in range(num_episodes):
        obs, _ = test_env.reset()
        done = False
        ep_reward = 0
        
        while not done:
            if agent_model:
                action, _ = agent_model.predict(obs, deterministic=True)
            else:
                action = test_env.action_space.sample() # ëœë¤ í–‰ë™
            
            obs, reward, done, _, info = test_env.step(action)
            ep_reward += reward
            
            # í†µê³„ ì§‘ê³„
            if info.get('action') == 1: check_counts['fobt'] += 1
            if info.get('action') == 2: check_counts['colo'] += 1
            
        rewards.append(ep_reward)
        lifespans.append(test_env.age)
        
        if info.get('cause') == 'cancer_death': deaths_cancer += 1
        if info.get('cause') == 'perforation_death': deaths_perf += 1

    avg_reward = np.mean(rewards)
    avg_life = np.mean(lifespans)
    
    print(f"í‰ê·  ë³´ìƒ: {avg_reward:.2f}")
    print(f"í‰ê·  ìˆ˜ëª…: {avg_life:.1f}ì„¸")
    print(f"ì•” ì‚¬ë§ë¥ : {(deaths_cancer/num_episodes)*100:.1f}%")
    print(f"í•©ë³‘ì¦ ì‚¬ë§: {(deaths_perf/num_episodes)*100:.1f}%")
    print(f"ì´ ê²€ì‚¬ëŸ‰: ë‚´ì‹œê²½ {check_counts['colo']}íšŒ, FOBT {check_counts['fobt']}íšŒ")
    return avg_reward

print("\nğŸ“Š í•™ìŠµ ì „ (Random) ì„±ëŠ¥ ì¸¡ì •...")
evaluate_agent(None, num_episodes=50, name="Random Agent")

# ==========================================
# 3. ëª¨ë¸ í•™ìŠµ (Training)
# ==========================================
print("\nğŸ”¥ í•™ìŠµ ì‹œì‘! (ì•½ 3~5ë¶„ ì†Œìš” ì˜ˆìƒ)")
# total_timesteps: ì‹œë®¬ë ˆì´ì…˜ ì´ íšŸìˆ˜ (ë…„ ë‹¨ìœ„)
# 20ë§Œ~30ë§Œ ë²ˆ ì •ë„ë©´ ê°„ë‹¨í•œ ì •ì±…ì€ ìˆ˜ë¦½ë©ë‹ˆë‹¤.
model.learn(total_timesteps=300000) 
print("âœ… í•™ìŠµ ì™„ë£Œ!")

# ëª¨ë¸ ì €ì¥
model.save("cmost_ppo_model")
print("ğŸ’¾ ëª¨ë¸ ì €ì¥ ì™„ë£Œ: cmost_ppo_model.zip")

# ==========================================
# 4. í•™ìŠµ í›„ ì„±ëŠ¥ í…ŒìŠ¤íŠ¸ (Evaluation)
# ==========================================
print("\nğŸ“Š í•™ìŠµ í›„ (PPO AI) ì„±ëŠ¥ ì¸¡ì •...")
evaluate_agent(model, num_episodes=50, name="Trained AI")

# ==========================================
# 5. ì‹œê°í™” (í•œ ëª…ì˜ í™˜ì ì¸ìƒ ë”°ë¼ê°€ê¸°)
# ==========================================
print("\nğŸ” [ìƒì„¸ ë¶„ì„] í•™ìŠµëœ AIì˜ ì§„ë£Œ ê¸°ë¡ (1ëª… ì˜ˆì‹œ)")
test_env = CmostEnv()
obs, _ = test_env.reset()
done = False
print(f"í™˜ì ì‹œì‘: {test_env.age}ì„¸ (ì—¬ì„±={test_env.is_female}, ìœ„í—˜ë„={test_env.risk_factor:.2f})")

history = []
while not done:
    action, _ = model.predict(obs, deterministic=True)
    obs, reward, done, _, info = test_env.step(action)
    
    act_str = ["ëŒ€ê¸°", "FOBT", "ë‚´ì‹œê²½"][action]
    if action > 0 or done: # ê²€ì‚¬í–ˆê±°ë‚˜ ì£½ì—ˆì„ ë•Œë§Œ ë¡œê·¸ ì¶œë ¥
        print(f"[{test_env.age}ì„¸] í–‰ë™: {act_str} | ë³´ìƒ: {reward:.1f} | ìƒíƒœ: {info['cause']}")
```

```python
#evaluation (vs random)

# --- [5] í•™ìŠµëœ ëª¨ë¸ ì„±ëŠ¥ í…ŒìŠ¤íŠ¸ ---

def run_simulation(agent_model, episodes=10):
    """ ì—ì´ì „íŠ¸ë¥¼ ì´ìš©í•´ í™˜ì ì‹œë®¬ë ˆì´ì…˜ì„ ëŒë¦¬ê³  ê²°ê³¼ë¥¼ ì¶œë ¥í•˜ëŠ” í•¨ìˆ˜ """
    # í…ŒìŠ¤íŠ¸ìš© í™˜ê²½ ë³„ë„ ìƒì„±
    test_env = CmostEnv()
    total_rewards = []
    lifespans = []
    deaths_by_cancer = 0
    
    print(f"\n--- ì‹œë®¬ë ˆì´ì…˜ ì‹œì‘ ({episodes}ëª… í™˜ì) ---")
    
    for ep in range(episodes):
        obs, _ = test_env.reset()
        done = False
        ep_reward = 0
        actions = []
        
        while not done:
            # ëª¨ë¸ì´ ìˆìœ¼ë©´ ì˜ˆì¸¡ê°’ ì‚¬ìš©, ì—†ìœ¼ë©´ ëœë¤ í–‰ë™
            if agent_model:
                action, _ = agent_model.predict(obs, deterministic=True)
            else:
                action = test_env.action_space.sample()
                
            obs, reward, done, _, info = test_env.step(action)
            ep_reward += reward
            actions.append(action)
            
        total_rewards.append(ep_reward)
        lifespans.append(test_env.age)
        if info.get('cause_of_death') == 'cancer':
            deaths_by_cancer += 1
            
        # ì²˜ìŒ 3ëª…ë§Œ ë¡œê·¸ ìì„¸íˆ ì¶œë ¥
        if ep < 3:
            print(f"í™˜ì {ep+1}: ì‚¬ë§ë‚˜ì´ {test_env.age}ì„¸, ì‚¬ì¸: {info.get('cause_of_death')}")
            # í–‰ë™ í†µê³„ (0:ëŒ€ê¸°, 1:FOBT, 2:ë‚´ì‹œê²½)
            unique, counts = np.unique(actions, return_counts=True)
            action_dict = dict(zip(unique, counts))
            print(f"   -> ê²€ì‚¬ íšŸìˆ˜: ë‚´ì‹œê²½ {action_dict.get(2, 0)}íšŒ, FOBT {action_dict.get(1, 0)}íšŒ")
            print(f"   -> ì´ ë³´ìƒ: {ep_reward:.2f}")

    print("-" * 30)
    print(f"í‰ê·  ìˆ˜ëª…: {np.mean(lifespans):.1f}ì„¸")
    print(f"í‰ê·  ë³´ìƒ: {np.mean(total_rewards):.2f}")
    print(f"ì•” ì‚¬ë§ë¥ : {(deaths_by_cancer/episodes)*100:.1f}%")

# 1. í•™ìŠµ ì „ (ëœë¤) ì„±ëŠ¥
print("\n[ë¹„êµ 1] í•™ìŠµ ì „ (Random Agent)")
run_simulation(None, episodes=100)

# 2. í•™ìŠµ í›„ (PPO) ì„±ëŠ¥
print("\n[ë¹„êµ 2] í•™ìŠµ í›„ (PPO Agent)")
run_simulation(model, episodes=100)
```

```python
import pandas as pd
import numpy as np

# íŒë‹¤ìŠ¤ ì¶œë ¥ ì œí•œ í•´ì œ (ì¤‘ìš”: ì´ ì„¤ì •ì„ í•´ì•¼ ì¤‘ê°„ ìƒëµì´ ì•ˆ ë©ë‹ˆë‹¤)
pd.set_option('display.max_rows', None)      # í–‰ ê°œìˆ˜ ì œí•œ í•´ì œ
pd.set_option('display.max_columns', None)   # ì—´ ê°œìˆ˜ ì œí•œ í•´ì œ
pd.set_option('display.width', 1000)         # ê°€ë¡œ ê¸¸ì´ ëŠ˜ë¦¬ê¸°
pd.set_option('display.max_colwidth', None)  # ì…€ ë‚´ìš© ì˜ë¦¼ ë°©ì§€

def get_full_patient_list(model, env_class, num_patients=100):
    env = env_class()
    records = []
    
    print(f"ğŸ” {num_patients}ëª…ì˜ í™˜ì ê¸°ë¡ ìƒì„± ì¤‘...", end="")
    
    for i in range(num_patients):
        if i % 10 == 0: print(".", end="") # ì§„í–‰ìƒí™© í‘œì‹œ
        
        obs, _ = env.reset()
        done = False
        
        # ê¸°ë¡ìš© ë³€ìˆ˜
        risk = env.risk_factor
        events = [] # (ë‚˜ì´, ê²€ì‚¬ì¢…ë¥˜)
        cause_of_death = None
        
        while not done:
            # AI í–‰ë™ ê²°ì •
            if model:
                action, _ = model.predict(obs, deterministic=True)
            else:
                action = env.action_space.sample()
            
            current_age = env.age
            
            # ê²€ì‚¬ ê¸°ë¡ (Action 1: FOBT, 2: Colo)
            if action == 1:
                events.append((current_age, "FOBT"))
            elif action == 2:
                events.append((current_age, "Colo"))
            
            obs, reward, done, _, info = env.step(action)
            
            if done:
                cause_of_death = info.get('cause')

        # --- ë°ì´í„° ê°€ê³µ ---
        # 1. ê²€ì‚¬ ì¢…ë¥˜ (Methods)
        methods = [e[1] for e in events]
        
        # 2. ê²€ì‚¬ ì£¼ê¸° (Intervals)
        test_ages = [e[0] for e in events]
        if len(test_ages) > 1:
            # ë‚˜ì´ ì°¨ì´ ê³„ì‚° (ì˜ˆ: 50ì„¸, 55ì„¸ -> 5ë…„ ê°„ê²©)
            intervals = np.diff(test_ages).tolist() 
            # ë³´ê¸° ì¢‹ê²Œ ì†Œìˆ˜ì  ì œê±° (5.0 -> 5)
            intervals = [int(x) for x in intervals]
        else:
            intervals = "-" # ê²€ì‚¬ê°€ 0íšŒ ë˜ëŠ” 1íšŒë¼ ê°„ê²© ì—†ìŒ

        # 3. ì‚¬ë§ ì›ì¸ í•œê¸€í™” (ë³´ê¸° ì¢‹ê²Œ)
        cause_map = {
            "natural_death": "ìì—°ì‚¬",
            "cancer_death": "ì•” ì‚¬ë§",
            "perforation_death": "ì˜ë£Œì‚¬ê³ (ì²œê³µ)",
            "max_age_reached": "ìˆ˜ëª…ì¢…ë£Œ(100ì„¸)",
            "cancer_detected": "ì•” ì™„ì¹˜(ë°œê²¬)"
        }
        cause_str = cause_map.get(cause_of_death, cause_of_death)

        # ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€
        records.append({
            "ID": i + 1,
            "Death Cause": cause_str,
            "Risk": round(risk, 2),
            "Total Tests": len(events),
            "Methods": str(methods).replace("'", ""), # ë”°ì˜´í‘œ ì œê±°í•´ì„œ ê¹”ë”í•˜ê²Œ
            "Intervals (Years)": str(intervals)
        })

    print(" ì™„ë£Œ!\n")
    return pd.DataFrame(records)

# --- ì‹¤í–‰ ë° ì¶œë ¥ ---
# 1. ë°ì´í„° ìƒì„±
df_full_list = get_full_patient_list(model, CmostEnv, num_patients=100)

# 2. ì „ì²´ ì¶œë ¥ (to_string()ì„ ì“°ë©´ ì˜ë¦¬ì§€ ì•Šê³  ë¬¸ìì—´ë¡œ ë‹¤ ë‚˜ì˜µë‹ˆë‹¤)
print(df_full_list.to_string(index=False))
```
